```python
import pandas as pd
```

```python
dev=pd.read_csv('/Users/paolanunez/Documents/TFG_PaolaNuñez/bert_repo/data/data_tab/val_posAneg_tab.csv',delimiter=",",header=None)
train=pd.read_csv('/Users/paolanunez/Documents/TFG_PaolaNuñez/bert_repo/data/data_tab/train_posAneg_tab.csv',delimiter=",",header=None)
```

```python
train1 = train.sample(frac=1)
dev1 = dev.sample(frac=1)
```

```python
train_df_bert = pd.DataFrame({
    'id':range(len(train1)),
    'label':train1[1],
    'alpha':[' ']*train1.shape[0],
    'text': train1[0].replace(r'\n', ' ', regex=True)
})
train_df_bert.head()
```

```python
dev_df_bert = pd.DataFrame({
    'guid':range(len(dev1)),
    'label':dev1[1],
    'alpha':[' ']*dev1.shape[0],
    'text': dev1[0].replace(r'\n', ' ', regex=True)
})

dev_df_bert.head()
```

```python
train_df_bert.to_csv('/Users/paolanunez/Documents/TFG_PaolaNuñez/bert_repo/data/data_tab/train.tsv', sep='\t', index=False, header=False)
dev_df_bert.to_csv('/Users/paolanunez/Documents/TFG_PaolaNuñez/bert_repo/data/data_tab/dev.tsv', sep='\t', index=False, header=False)
```
