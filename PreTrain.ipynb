```python
import tokenizers
bwpt = tokenizers.BertWordPieceTokenizer(
    vocab_file=None,
    unk_token='[UNK]',
    sep_token='[SEP]',
    cls_token='[CLS]',
    clean_text=True,
    handle_chinese_chars=False,
    strip_accents=True,
    lowercase=True,
    wordpieces_prefix='##'
)
```

```python
bwpt.train(
    files=["/Users/paolanunez/Documents/TFG_PaolaNuñez/bert_repo/data/data_tab/train_posAneg_tab3.txt"],
    vocab_size=27,
    min_frequency=3,
    limit_alphabet=50,
    special_tokens=['[PAD]', '[UNK]', '[CLS]', '[MASK]', '[SEP]']
)
```

```python
bwpt.save("/Users/paolanunez/Documents/TFG_PaolaNuñez/bert_repo/model_tab", "protein_tab")
```
